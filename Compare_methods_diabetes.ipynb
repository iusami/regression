```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
import scipy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

diabetes = datasets.load_boston()
```

```python
dframe = pd.DataFrame(diabetes.data)
dframe.columns = diabetes.feature_names
#dframe.columns = "age", "sex", "bmi", "map", "tc", "ldl", "hdl", "tch", "ltg", "glu"
dframe["y"] = diabetes.target
print(dframe.head())
```

```python
#get BMI data and modificate it to ndarray normalized
#diabetes_X = scipy.stats.zscore(diabetes.data[:,np.newaxis,12])
#diabetes_Y = scipy.stats.zscore(diabetes.target)
diabetes_X = scipy.stats.zscore(dframe.loc[:,["LSTAT"]].values)
diabetes_Y = scipy.stats.zscore(dframe.loc[:,"y"].values)


#get BMI data and modificate it to ndarray
#diabetes_X = diabetes.data[:,np.newaxis,12]
#diabetes_Y = diabetes.target
#diabetes_X = dframe.loc[:,["LSTAT"]].values
#diabetes_Y = dframe.loc[:,"y"].values

#print(diabetes_X)
#diabetes_X_panda = dframe["bmi"]
#print(np.reshape(diabetes_X_panda.values,[diabetes_X_panda.values.shape[0],1]).shape)

#split data into training/test
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]
diabetes_Y_train = diabetes_Y[:-20]
diabetes_Y_test = diabetes_Y[-20:]

#sb.pairplot(dframe)
df_standarized = pd.DataFrame(scipy.stats.zscore(dframe),
                              index=dframe.index,columns=dframe.columns)
sb.pairplot(dframe[["LSTAT","y"]])
```

```python


#create model
regre = linear_model.LinearRegression()
ridge = linear_model.Ridge(alpha=.5)
lasso =  linear_model.Lasso(alpha=0.1)
regre_quad = linear_model.LinearRegression()
regre_cubic = linear_model.LinearRegression()

#polynomial instance
quadratic = PolynomialFeatures(degree=2)
cubic = PolynomialFeatures(degree=3)
X_quad = quadratic.fit_transform(diabetes_X_train)
X_cubic = cubic.fit_transform(diabetes_X_train)

# モデル式用に変数を作成
X_fit = np.arange(diabetes_X_train.min(), diabetes_X_train.max(), 1)[:, np.newaxis]

#Train model
regre.fit(diabetes_X_train, diabetes_Y_train)
ridge.fit(diabetes_X_train, diabetes_Y_train)
lasso.fit(diabetes_X_train, diabetes_Y_train)
regre_quad.fit(X_quad, diabetes_Y_train)
regre_cubic.fit(X_cubic, diabetes_Y_train)

#Make predictioins using train data
regre_lin = regre.predict(X_fit)
regre_quad_fit = regre_quad.predict(quadratic.fit_transform(X_fit))
regre_cubic_fit = regre_cubic.predict(cubic.fit_transform(X_fit))

#Make predictions using the test data
diabetes_y_pred = regre.predict(diabetes_X_train)
ridge_y_pred = ridge.predict(diabetes_X_test)
lasso_y_pred = lasso.predict(diabetes_X_test)
quad_y_pred = regre_quad.predict(quadratic.fit_transform(diabetes_X_train))
cubic_y_pred = regre_cubic.predict(cubic.fit_transform(diabetes_X_train))

#print(diabetes_y_pred)
#coefficients
#回帰係数。y=ax+bのaにあたる
print("Coefficiens:",regre.coef_)
print("Coefficiens_ridge:",ridge.coef_)
print("Coefficiens_lasso:",lasso.coef_)
print("Coefficiens_quad:",regre_quad.coef_)
print("Coefficiens_cubic:",regre_cubic.coef_)


#Mean Squared Error
#平均二乗誤差。小さければ性能が良い。
"""
print("Mean Squared Error:",mean_squared_error(diabetes_Y_test,diabetes_y_pred))
print("Mean Squared Error_ridge:",mean_squared_error(diabetes_Y_test,ridge_y_pred))
print("Mean Squared Error_lasso:",mean_squared_error(diabetes_Y_test,lasso_y_pred))
print("Mean Squared Error_quad:",mean_squared_error(diabetes_Y_test,quad_y_pred))
print("Mean Squared Error_cubic:",mean_squared_error(diabetes_Y_test,cubic_y_pred))
"""
#R^2
#決定係数。１に近いほど性能が良い。
print("R^2:",r2_score(diabetes_Y_train,diabetes_y_pred))
print("R^2_ridge:",r2_score(diabetes_Y_test,ridge_y_pred))
print("R^2_lasso:",r2_score(diabetes_Y_test,lasso_y_pred))
print("R^2_quad:",r2_score(diabetes_Y_train,quad_y_pred))
print("R^2_cubic:",r2_score(diabetes_Y_train,cubic_y_pred))
```

```python
# Plot outputs
#plt.scatter(diabetes_X_test, diabetes_Y_test,  color='black')
plt.scatter(diabetes_X_train, diabetes_Y_train,  color='lightgray')
plt.plot(X_fit, regre_lin, color='blue', linewidth=1,label="linear")
#plt.plot(diabetes_X_test, ridge_y_pred, color='red', linewidth=3)
#plt.plot(diabetes_X_test, lasso_y_pred, color='green', linewidth=3)
plt.plot(X_fit, regre_quad_fit, color='red', 
         linewidth=1,label="quad",linestyle = "-")
plt.plot(X_fit, regre_cubic_fit, color='green', 
         linewidth=1,label="cubic",linestyle = "--")

#plt.xticks(())
#plt.yticks(())
plt.legend()
plt.show()
```
